```python
# ---------------------------- NAN VALS EDA -----------------------------------
df.isna().sum() #theres missing vals in the n_gxs and hospital_rate cols
```

```python
# checking the missing in certain months
df[df['n_gxs'].isna()].groupby('months_postgx').size()
```

```python
df[(df['months_postgx'] == 0) & (df['n_gxs'].isna())].shape[0] # checking if there are missing n_gxs col in month 0
```

The above shows that 1953 missing values are missing at every pre-generic-entry month (-24 to -1 before generics). So all drugs prior to generic entry have no generics which is correct, however this means that all NaN vals within the -24 to -1 month period should be imputed with 0, representing 0 generics made.\n

After post-generic-entry (months 0+) there are 514 missing vals due to the merging, the df_generics dataset contains only the drugs that record change where there is a new generic entry, so for every change in number of generic entry it is recorded in df_generics provided dataset. Meaning that the missing vals post-entry just mean that there is no change within that specific month, thus it should be imputed with the previous month's value.\n

At month 0 that is where erosion begins and this is where our model begins meaning at least 1 generic entry needs to exist at month 0.

## Data Cleaning

### NaN Values

```python
# pre generic entry (months -24 to -1), there should be 0 generics as per provided documentation
df.loc[df['months_postgx'] < 0, 'n_gxs'] = 0

# post generic entry months (0+), should be the same as previous month (explained why above)
df['n_gxs'] = df.groupby(['country','brand_name'])['n_gxs'].ffill()
df['n_gxs'] = df['n_gxs'].astype(int)
```
