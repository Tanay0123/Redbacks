```python
# load the data
import pandas as pd
train_data = pd.read_csv("Data/TRAIN/merged_cleaned_train_data.csv")
test_data = pd.read_csv("Data/TEST/novartis_test_data_merged_full(New).csv")
```

```python
train_data = train_data.drop(columns=["Unnamed: 0"])
```

```python
train_data.info()
```

```python
test_data.info()
```

```python
# identify input and output
X_train = train_data.drop("volume", axis=1)
X_test = test_data.drop("volume", axis=1)
y_train = train_data.volume
y_test = test_data.volume
```

```python
# one hot encode
from sklearn.preprocessing import OneHotEncoder
enc_train = OneHotEncoder()
enc_train.fit(X_train)
X_train_enc = enc_train.transform(X_train)

enc_test = OneHotEncoder()
enc_test.fit(X_test)
X_test_enc = enc_test.transform(X_test)
```

```python
# standardise
from sklearn.preprocessing import StandardScaler
std = StandardScaler(with_mean=False)
X_train_std = std.fit_transform(X_train_enc)
X_test_std = std.fit_transform(X_test_enc)
```

```python
# initialise and train the Poisson Regression model
from sklearn.linear_model import PoissonRegressor
poisson_model = PoissonRegressor()
poisson_model.fit(X_train_std, y_train)
```

```python
# Make predictions on the test set
y_pred = poisson_model.predict(X_test_std)
```
